### running code for all 10 companies

```python
window_size = 20
step_size = 10
```

```python
gaf = GramianAngularField(image_size=20, method='difference')
```

```python
def ts_to_gaf(images_df, cmp_name, window_size, step_size):

  df = pd.read_csv(root_path+'ten_companies_cleaned_csvs/'+cmp_name+'.csv')

  for i in tqdm(range(0, len(df) - window_size + 1, step_size), leave=False):

    subset = df.iloc[i:i+window_size]

    start_date = subset['Date'].iloc[0]
    end_date = subset['Date'].iloc[-1]
    prediction = 1 if subset['Close'].iloc[-1] < df.iloc[i+window_size]['Close'] else 0
    segment = np.array(subset['Close'].to_list()).reshape(1, -1)

    image_path = f'/content/drive/MyDrive/Colab Notebooks/CS583 Project/GAF/GAF_img_data/{cmp_name}_gaf_{i}.png'

    gaf_image = gaf.fit_transform(segment)

    plt.figure(figsize=(8, 8))
    plt.imshow(gaf_image[0], cmap='rainbow', origin='lower')
    plt.tight_layout()
    plt.axis('off')
    plt.subplots_adjust(left=0, right=1, top=1, bottom=0)
    plt.savefig(image_path)
    plt.close()

    images_df = images_df.append({
            'Company' : cmp_name,
            'Image_Path': image_path,
            'Start_Date': start_date,
            'End_Date': end_date,
            'Prediction': prediction
        }, ignore_index=True)

  return images_df
```

```python
images_df = pd.DataFrame(columns=['Company', 'Image_Path', 'Start_Date', 'End_Date', 'Prediction'])

for cmp_name in tqdm(ten_companies):
    images_df = ts_to_gaf(images_df, cmp_name, window_size, step_size)
```
